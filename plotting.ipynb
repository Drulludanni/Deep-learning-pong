```python

import matplotlib.pyplot as plt
import pickle
import os
# named the files .loss even though what they contain is the reward
rewards = [i for i in os.listdir() if ".loss" in i ]

reward_dict = {}
for i in rewards:
    with open( i, "rb" ) as file:
        reward_dict[i] = pickle.load(file)
rewards
```

```python
def plot_rewards(names, legend = None):
    plt.title('Training...')
    plt.xlabel('Episode')
    plt.ylabel('Rewards')
    for n in names:
        plt.plot(reward_dict[n])
    if legend is None:
        legend = names
    plt.legend(legend)
    plt.axhline(y=0.0,color='black',linestyle='--')
    plt.show()
mem_size = [
    'basic_10K_replay.loss',
    'basic_50K_replay.loss',
    'basic_100K_replay.loss']
plot_rewards(mem_size, ["10k replay", "50k replay", "100k replay"])
    
```

```python
prioritized = [
    'basic_100K_replay.loss',
    'basic_50K_replay.loss',
    'prioritized_50K_replay_2s.loss',
    'prioritized_100K_replay.loss',
]
plot_rewards(prioritized, ["100k replay", "50k replay", "prioritized 50k replay", "prioritized 100k replay"])
```

```python
soft = [
    'prioritized_100K_replay.loss',
    'prioritized_100K_replay_soft_updates.loss',
]
plot_rewards(soft,["every 1000", "soft updates"])
```

```python
deterministic = [
    'prioritized_100K_replay.loss',
    'deterministic_prioritized_100K_replay.loss',
]
plot_rewards(deterministic, ["nondeterministic","deterministic"])
```
